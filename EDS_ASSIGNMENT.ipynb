```python
import pandas as pd
df = pd.read_csv("/content/matches.csv")
```

(1) How many matches were played in total?

```python
total_matches = len(df)
print(total_matches)
```

(2)  How many unique seasons are represented in the dataset?

```python
unique_seasons = df['season'].nunique()
print(unique_seasons)
```

(3)  Which team has won the most matches?

```python
most_wins = df['winner'].value_counts().idxmax()
print(most_wins)
```

(4) How many matches were won by runs vs by wickets?

```python
wins_by_runs = (df['win_by_runs'] > 0).sum()
wins_by_wickets = (df['win_by_wickets'] > 0).sum()
print(wins_by_runs)
print(wins_by_wickets)
```

(5) Which city hosted the most matches?

```python
top_city = df['city'].value_counts().idxmax()
print(top_city)
```

(6)  How many matches had Duckworth-Lewis (DL) applied?

```python
dl_matches = df['dl_applied'].sum()
print(dl_matches)
```

(7) Which player won the most 'Player of the Match' awards?

```python
top_player = df['player_of_match'].value_counts().idxmax()
print(top_player)
```

(8) Which team won the toss most often?

```python
top_toss_team = df['toss_winner'].value_counts().idxmax()
print(top_toss_team)
```

(9) What percentage of toss-winning teams chose to field first?

```python
field_pct = (df['toss_decision'] == 'field').mean() * 100
print(field_pct)
```

(10) Which match had the highest win by runs?

```python
highest_run_win = df[df['win_by_runs'] == df['win_by_runs'].max()]
print(highest_run_win)
```

(11) Which match had the highest win by wickets?

```python
highest_wicket_win = df[df['win_by_wickets'] == df['win_by_wickets'].max()]
print(highest_wicket_win)
```

(12) Which season had the most matches played?

```python
most_matches_season = df['season'].value_counts().idxmax()
print(most_matches_season)
```

(13) How many matches were tied?

```python
tied_matches = (df['result'] == 'tie').sum()
print(tied_matches)
```

(14) What is the average margin of victory by runs?

```python
avg_run_win = df[df['win_by_runs'] > 0]['win_by_runs'].mean()
print(avg_run_win)
```

(15) What is the average margin of victory by wickets?

```python
avg_wicket_win = df[df['win_by_wickets'] > 0]['win_by_wickets'].mean()
print(avg_wicket_win)
```

(16) List the top 5 most frequent venues.

```python
top_venues = df['venue'].value_counts().head(5)
print(top_venues)
```

(17) How many matches had missing umpire1 or umpire2 data?

```python
missing_umpires = df[['umpire1', 'umpire2']].isnull().any(axis=1).sum()
print(missing_umpires)
```

(18) How many different teams participated in total?

```python
unique_teams = pd.unique(df[['team1', 'team2']].values.ravel('K'))
print(unique_teams)
```

(19) Which team has the best toss-to-win conversion rate?

```python
toss_wins = df[df['toss_winner'] == df['winner']]['winner'].value_counts()
toss_total = df['toss_winner'].value_counts()
conversion_rate = (toss_wins / toss_total).sort_values(ascending=False)
print(toss_wins)
print(toss_total)
print(conversion_rate)
```

(20) Are there any matches with missing winner information?

```python
missing_winner = df['winner'].isnull().sum()
print(missing_winner)
```
